## Imports

```python
from arc import *
import numpy as np
%matplotlib inline
```

```python
atom = Rubidium87()
nmin = 5  # Minimum n
nmax = 42  # Maximum n
lmin = 0  # Minimum l
lmax = 2  # Maxmium l

# Plot Energy Levels of Cesium
levels = LevelPlot(atom)
levels.makeLevels(nmin, nmin, 0, 1)
levels.makeLevels(nmax, nmax,2,2)
# levels.makeLevels(0,2,0,0)
levels.drawLevels()
levels.showPlot()
```

```python
ground_state = atom.getEnergy(n=5,l=0,j=+0.5)
excited_state = atom.getEnergy(n=5,l=1,j=+1.5)
rydberg_state = atom.getEnergy(n=42,l=2,j=+2.5)

ground_state, excited_state, rydberg_state

# excited_state - ground_state
```

```python
Gamma = atom.getTransitionRate(5, 1, 1.5, 5, 0, 0.5)
Gamma2 = atom.getTransitionRate(42,2,2.5, 5, 1, 1.5)
Gamma3 = atom.getTransitionRate(42,2,2.5, 5, 0, 0.5)
# print(Gamma2,Gamma, Gamma3)

Gamma2/Gamma
```

```python
probe_power = 1e-9
probe_waist = 5e-6

couple_power = 20e-3
couple_waist = 5e-6

Omega_p = atom.getRabiFrequency(5, 0, 0.5, 0.5, 5, 1, 1.5, 1, probe_power, probe_waist)/(2*np.pi)
Omega_c = atom.getRabiFrequency(5, 1, 1.5, 1.5, 42, 2, 2.5, 1, couple_power, couple_waist)/(2*np.pi)

print(Omega_p/Gamma)
print(Omega_c/Gamma)
# print(Omega_p)
# print(Omega_c)
```
